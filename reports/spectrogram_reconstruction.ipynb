```python
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = "/"+os.path.join("cluster","home","llorenz","sleep","MT18_LH_human-sleep-classification","logs","fold0","spectograms.np.npy")
file = "/"+os.path.join("cluster","home","llorenz","sleep","MT18_LH_human-sleep-classification","logs","cv_ready","caro","states_usv_h128f128","fold0","spectograms.np.npy")

sp = np.load(file).item()
```

```python
print(sp.keys())
truth = sp["truth"]
rec = sp["reconstruction"]
print(len(truth))
print(len(rec))
print(truth[0].shape, rec[0].shape)
```

```python
for i in range(len(truth)):
    t = truth[i][0,4]
    r = rec[i][0,4]
    
    n_plt = 2
        
    fig, axarr = plt.subplots(n_plt, sharex=True, sharey=False, figsize=(20, 20))
        
    pc1 = axarr[0].pcolormesh(t, cmap='jet', vmin=-5)
    #axarr[0].set_ylim(ymax=40)
    axarr[0].invert_yaxis()
    axarr[0].set_ylabel("spectogram [Hz]")
    axarr[0].figure.colorbar(pc1, ax=axarr[0])

    pc2 = axarr[1].pcolormesh(r, cmap='jet', vmin=-5)
    axarr[1].invert_yaxis()
    axarr[1].set_ylabel("spectogram [Hz]")
    axarr[1].figure.colorbar(pc2, ax=axarr[1])
```
