# We analyze the problem of classifying sleep phases

### (1) We start by importing necessary pre-installed packages

```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
import glob
```

### (2) We import our 'sleeplearning' library

```python
root_dir = os.path.abspath(os.path.join(os.path.dirname('__file__'), '..'))
sys.path.insert(0, root_dir)
from sleeplearning.lib.base import SleepLearning
from sleeplearning.lib.visualize import Visualize
%load_ext autoreload
%autoreload 2
```

### (3) We load all the mat files from a folder

```python
caro_subjects = []
for i, filename in enumerate(glob.glob(root_dir + '/data/caroline/*.mat')):
    mat = SleepLearning.read_caro(filename)
    caro_subjects.append(mat)
    print('')
```

### (4) We plot some overall statistics

```python
vcaro = Visualize(caro_subjects)
vcaro.class_distribution()
```

```python
vcaro.transition_distribution()
```
